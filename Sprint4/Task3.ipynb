Описание задачи:

Создать базовый класс ShippingOption в C#, который будет представлять различные опции доставки. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу ShippingOption:

• Атрибуты: ID опции доставки (DeliveryOptionId), Название опции доставки (DeliveryOptionName), Стоимость доставки (Cost).

• Методы:

o CalculateCost(): метод для расчета стоимости доставки.

o EstimateDeliveryTime(): метод для оценки времени доставки.

o GetDeliveryDetails(): метод для получения деталей опции доставки.

Требования к производным классам:

СтандартнаяДоставка (StandardDelivery): Должна содержать дополнительные атрибуты, такие как Среднее время доставки (AverageDeliveryTime). Метод EstimateDeliveryTime() должен быть переопределен для предоставления среднего времени доставки.

ЭкспрессДоставка (ExpressDelivery): Должна содержать дополнительные атрибуты, такие как Минимальное время доставки (MinDeliveryTime). Метод CalculateCost() должен быть переопределен для увеличения стоимости доставки в случае необходимости ускорения доставки.

Самовывоз (Pickup) (если требуется третий класс): Должна содержать дополнительные атрибуты, такие как Адрес пункта самовывоза (PickupAddress). Метод GetDeliveryDetails() должен быть переопределен для отображения адреса пункта самовывоза.

```
//интерфейс
public interface IInsurable
{
    bool IsInsured { get; } //Свойство, доступное только для чтения
}

//Интерфейс, который наследует IInsurable
public interface IDelivery : IInsurable
{
    string EstimateDeliveryTime(); 
    void GetDeliveryDetails();
}

//Класс, реализующий интерфейс 
public class ShippingOption : IDelivery
{

    
    public int DeliveryOptionId { get; } 
    public string DeliveryOptionName { get; }
    public double Cost { get; }

   
    public ShippingOption(int id, string name, double cost, bool isInsured)
    {
        DeliveryOptionId = id; 
        DeliveryOptionName = name; 
        Cost = cost;
        IsInsured = isInsured; //Устанавливаем значение для свойства
    }

    //Явная реализация свойства из интерфейса IInsurable
    public bool IsInsured { get; } 


    public virtual string EstimateDeliveryTime()
    {
        return "Срок доставки неизвестен."; 
    }


    public virtual void GetDeliveryDetails()
    {
        //Вывод информации о доставке, включая значение страховки через явную реализацию
        Console.WriteLine($"ID доставки: {DeliveryOptionId}, Тип: {DeliveryOptionName}, Стоимость: {Cost}Р, Страховка: {((IInsurable)this).IsInsured}");
    }
}

//Класс, наследующий ShippingOption
public class StandardDelivery : ShippingOption
{
    public int AveragePickupTimeDays { get; } 
    public string Destination { get; } 

    
    public StandardDelivery(int id, string name, double cost, int averagePickupTimeDays, string destination, bool isInsured)
        : base(id, name, cost, isInsured) 
    {
        AveragePickupTimeDays = averagePickupTimeDays; 
        Destination = destination; 
    }

    //Переопределение метода 
    public override string EstimateDeliveryTime()
    {
        return $"Ожидаемое время доставки: {AveragePickupTimeDays} дней."; 
    }

    public override void GetDeliveryDetails()
    {
        base.GetDeliveryDetails(); 
        Console.WriteLine($"Место назначения: {Destination}"); 
    }
}


public class ExpressDelivery : ShippingOption
{
    public double AveragePickupTimeHours { get; } 
    public string Destination { get; } 

    public ExpressDelivery(int id, string name, double cost, double averagePickupTimeHours, string destination, bool isInsured)
        : base(id, name, cost, isInsured) // Вызов конструктора базового класса
    {
        AveragePickupTimeHours = averagePickupTimeHours; 
        Destination = destination; 
    }

    public override string EstimateDeliveryTime()
    {
        return $"Ожидаемое время доставки: {AveragePickupTimeHours} часов."; 
    }

    public override void GetDeliveryDetails()
    {
        base.GetDeliveryDetails(); // Вызов метода базового класса
        Console.WriteLine($"Место назначения: {Destination}");
    }
}


public class Pickup : ShippingOption
{
    public int AveragePickupTimeDays { get; } 
    public string PickupLocation { get; } 

    public Pickup(int id, string name, double cost, int averagePickupTimeDays, string pickupLocation, bool isInsured)
        : base(id, name, cost, isInsured) // Вызов конструктора базового класса
    {
        AveragePickupTimeDays = averagePickupTimeDays; 
        PickupLocation = pickupLocation; 
    }

    public override string EstimateDeliveryTime()
    {
        return $"Ожидаемое время самовывоза: {AveragePickupTimeDays} дней."; 
    }

    public override void GetDeliveryDetails()
    {
        base.GetDeliveryDetails(); 
        Console.WriteLine($"Место самовывоза: {PickupLocation}"); 
    }
}


public class StandardPickup : Pickup
{
    public TimeSpan AverageStorageTime { get; } 

    // Конструктор класса StandardPickup
    public StandardPickup(int id, string name, double cost, int averagePickupTimeDays, string pickupLocation, bool isInsured, TimeSpan averageStorageTime)
        : base(id, name, cost, averagePickupTimeDays, pickupLocation, isInsured) // Вызов конструктора базового класса
    {
        AverageStorageTime = averageStorageTime;
    }

    public override void GetDeliveryDetails()
    {
        base.GetDeliveryDetails(); 
        Console.WriteLine($"Срок хранения доставки: {AverageStorageTime.TotalDays} дней."); 
    }
}

//Класс для управления доставкой, наследующийся от интерфейса IDelivery
public class DeliveryService
{
    private readonly IDelivery _deliveryOption; //Поле для хранения варианта доставки (управление зависимостями)

    //Конструктор класса 
    public DeliveryService(IDelivery deliveryOption)
    {
        _deliveryOption = deliveryOption; //Установка зависимости через конструктор
    }

    //Метод для отображения деталей доставки
    public void DisplayDeliveryDetails()
    {
        _deliveryOption.GetDeliveryDetails(); 
        Console.WriteLine(_deliveryOption.EstimateDeliveryTime());
    }
}

IDelivery standardDelivery = new StandardDelivery(1, "Стандартная доставка", 500, 2, "Тобольск", true);
IDelivery expressDelivery = new ExpressDelivery(2, "Экспресс доставка", 950, 3, "Омутинское", false);
IDelivery pickup = new Pickup(3, "Самовывоз", 0, 5, "Тюмень", false);
IDelivery standardPickup = new StandardPickup(4, "Стандартный самовывоз", 0, 2, "Талица", false, TimeSpan.FromDays(2));

DeliveryService deliveryService1 = new DeliveryService(standardDelivery); // Передаем зависимость
deliveryService1.DisplayDeliveryDetails(); // Отображаем детали
Console.WriteLine();

DeliveryService deliveryService2 = new DeliveryService(expressDelivery); 
deliveryService2.DisplayDeliveryDetails(); 
Console.WriteLine();

DeliveryService deliveryService3 = new DeliveryService(pickup); 
deliveryService3.DisplayDeliveryDetails(); 
Console.WriteLine();

DeliveryService deliveryService4 = new DeliveryService(standardPickup); 
deliveryService4.DisplayDeliveryDetails();
```
